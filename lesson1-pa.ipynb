```python
import pandas as pd
```

```python
authors = pd.DataFrame({'author_id':[1, 2, 3], 'author_name':['Тургенев', 'Чехов', 'Островский']}, 
                       columns=['author_id', 'author_name'])
```

```python
authors
```

```python
book = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3], 
                     'book_title':['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 
                                   'Гроза', 'Таланты и поклонники'],
                     'price':['450', '300', '350', '500', '450', '370', '290']}, columns=['author_id', 'book_title', 'price'])
```

```python
book
```

```python
authors_price = pd.merge(authors, book, on = 'author_id', how = 'outer')
authors_price
```

```python
top5 = authors_price.sort_values(by='price').tail(5)
top5
```

или так:

```python
authors_price['price'] = authors_price['price'].astype(int)
top5 = authors_price.nlargest(5, 'price')
top5
```

```python
authors_stat = authors_price.groupby('author_name').agg(price_min = ('price', 'min'), price_max = ('price', 'max'), 
                                                       price_mean = ('price', 'mean'))
authors_stat['price_mean'] = round(authors_stat['price_mean'], 2)
authors_stat
```

```python
cover = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price['cover'] = cover
authors_price
```

```python
book_info = authors_price.pivot_table('price', index='author_name', columns='cover', aggfunc='sum', fill_value = 0)
book_info
```

```python
book_info.to_pickle('book_info')
book_info_2 = pd.read_pickle('book_info')
book_info_2
```
