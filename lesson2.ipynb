```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, c = 'g')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title('График f(t)', fontsize=18, fontweight='bold', color='green', family='calibri')
plt.xlabel('значение t')
plt.ylabel('значение y')
```

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
title_dict = {'fontsize': '15', 'fontweight': 'bold', 'color': 'green', 'family': 'calibri'}

fig, ax = plt.subplots(nrows = 2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1', fontdict=title_dict)
ax1.grid()
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2', fontdict=title_dict)
ax2.grid()

ax3.plot(x, y3)
ax3.set_title('График y3', fontdict=title_dict)
ax3.grid()

ax4.plot(x, y4)
ax4.set_title('График y4', fontdict=title_dict)
ax4.grid()
# можно ли прописать одной строкой, чтобы сетка отображалась на каждом из графиков?
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
